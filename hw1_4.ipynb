```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os

plt.rcParams['figure.figsize'] = [16, 8]

mu = 0 
sd = 1 #normal distribution
a = np.random.normal(mu, sd, size=(100, 100)) #random 100x100 matrix

U, S, VT = np.linalg.svd(a,full_matrices=True) #compute the full SVD
S = np.diag(S)

plt.figure(1)  #plot the singular values 
plt.semilogy(np.diag(S))
plt.title('Singular values')
plt.show()
```

```python
#Repeat this 100 times and plot the distribution of the singular values in a box-and-wisker plot 

data = []

for i in range(1,100):
    b = np.random.normal(mu, sd, size=(100, 100)) #random 100x100 matrix
    U_b, S_b, VT_b = np.linalg.svd(b,full_matrices=True) #compute the full SVD
    S_b = S_b.tolist()
    data.append(S_b)
    
```

```python
data = np.array(data)
data = np.swapaxes(data,0,1)
#print(np.shape(data))
fig = plt.figure(figsize =(16, 8))
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(data)
plt.show()
```

```python
mean_tot = np.empty(99)
median_tot = np. empty(99)

for i in range (1,100):
    mean_tot[i-1] = np.mean(data[:,i-1])
    median_tot[i-1] = np.median(data[:,i-1])
```

```python
plt.figure(3)
plt.subplot(121)
plt.plot(mean_tot)
plt.title('Mean')
plt.subplot(122)
plt.plot(median_tot)
plt.title('Median')
plt.show()
```

```python
# repeat for matrix size 50x50

c = np.random.normal(mu, sd, size=(50, 50)) #random 50x50 matrix

U_c, S_c, VT_c = np.linalg.svd(c,full_matrices=True) #compute the full SVD
S_c = np.diag(S_c)

plt.figure(4)  #plot the singular values 
plt.semilogy(np.diag(S_c))
plt.title('singular values')
plt.show()

data_c = []

for i in range(1,50):
    d = np.random.normal(mu, sd, size=(50, 50)) #random 100x100 matrix
    U_d, S_d, VT_d = np.linalg.svd(d,full_matrices=True) #compute the full SVD
    S_d = S_d.tolist()
    data_c.append(S_d)
    
    
data_c = np.array(data_c)
data_c = np.swapaxes(data_c,0,1)

fig = plt.figure(figsize =(16, 8))
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(data_c)
plt.show()
```
