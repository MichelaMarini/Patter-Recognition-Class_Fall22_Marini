```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
import random
plt.rcParams['figure.figsize'] = [16, 8]

throw = 10000
L = 2
j = 0

for i in range(throw):
    x = random.uniform(0,L)
    y = random.uniform(0,L)
    
    if x*x + y*y < 1:
      j += 1
        
        
print("number of radius < 1:", j)
```

```python
k = np.zeros(9)
l = 0


for N in range(2,10):
    radius = []
    
    for i in range(throw):  
        x = np.random.uniform(0,L,N)
        x = np.reshape(x, (-1,N))
        r = x@x.T
    
        if r[0,0] < 1: 
          radius.append(r[0,0])
          k[l] += 1
          
    l += 1
    #print(radius)
    
    plt.hist(radius)
    plt.title('N = '+str(N))
    plt.show()
        
        
print("number of radius < 1, N=2:", k[0])
print("number of radius < 1, N=3:", k[1])
print("number of radius < 1, N=4:", k[2])
print("number of radius < 1, N=5:", k[3])
print("number of radius < 1, N=6:", k[4])
print("number of radius < 1, N=7:", k[5])
print("number of radius < 1, N=8:", k[6])
print("number of radius < 1, N=9:", k[7])
print("number of radius < 1, N=10:", k[8])
```

```python
N_1 = np.array([2,3,4,5,6,7,8,9,10])
plt.plot(N_1,k)
plt.xlabel("N dimension")
plt.ylabel("# of points land inside the sphere")
plt.show()
```
