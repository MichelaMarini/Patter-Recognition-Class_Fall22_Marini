```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
import random

from matplotlib.image import imread
from scipy import sparse
from scipy.fftpack import dct, idct
from scipy.optimize import minimize
from cosamp_fn import cosamp

plt.rcParams['figure.figsize'] = [16, 8]


n= 1000
K = 5

s = np.zeros(n)
index= np.floor(np.random.rand(K) * n).astype(int)
value= np.floor(np.random.rand(K) * 10).astype(int)
s[index]= value
#K sparse vector s of Fourier coefficients
```

```python
## Randomly sample signal
avg_error = []
for p in range (1,100): #100
    Error = []
    error = np.empty(10)
    for i in range(10):
        C = np.random.randn(p,n)
        Psi = dct(np.identity(n))
        Theta = C @ Psi
        y_true = Theta @ s
        s_hat = cosamp(Theta,y_true,p) # CS via matching pursuit
        error[i] = np.linalg.norm(s - s_hat) / np.linalg.norm(s)
    error_10 = np.mean(error)
    avg_error.append(error_10)
    
```

```python
l, ax = plt.subplots(figsize=(16, 8))
ax.plot(np.arange(1,100),avg_error)
ax.set_xlabel('p')
ax.set_ylabel('error')
ax.set_title('Average Relative Error')  
```
